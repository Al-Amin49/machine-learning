```python
"""
Dataset: The annual salaries (in $1000) of 20 employees in a tech company:
[45, 48, 50, 52, 55, 58, 60, 62, 65, 68, 70, 72, 75, 78, 80, 85, 90, 95, 150, 500]

- Calculate:
- Range
- Variance 
- Standard deviation 
- Q2 
- IQR (Interquartile Range)
- Calculate the lower and upper fences for outlier detection
- Identify any outliers using the IQR method
- How many outliers are there and what are their values?

"""
```

```python
import numpy as np
data=[45, 48, 50, 52, 55, 58, 60, 62, 65, 68, 70, 72, 75, 78, 80, 85, 90, 95, 150, 500]

data_np=np.array(data)
range=data_np.max()-data_np.min()
variance=np.var(data_np)
std=np.std(data_np)
Q1=np.percentile(data_np, 25)
Q2=np.percentile(data_np, 50)
Q3=np.percentile(data_np, 75)
IQR=Q3-Q1
lower_bound= Q1-(1.5*IQR)
upper_bound= Q3+(1.5*IQR)
identify_outliers=data_np[(data_np < lower_bound) | (data_np >upper_bound)]
#range
print(f"Range: {range} = {data_np.max()} - {data_np.min()}")
#variance
print(f"Variance : {variance:.2f}")
#standard deviation
print(f"STD : {std:.2f}")
# Q2
print("Q2: ", Q2 )
#IQR
print("IQR : ", IQR)
print("Lower fences ", lower_bound )
print("Upper fences: ", upper_bound)
#identify outliers
print(f"Outliers: {identify_outliers}")
print(f"Number of outliers: {len(identify_outliers)}")
```
