### Standardization (z-score scaling)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.DataFrame({'h':[150,160,170,185,190],
                   'w':[50,58,70,80,90]})
# list(df.index)
```

```python
#*Formula z=(x-mean)/standard deviation
mean_val=df.mean()
s=df.std()
print(s)
z=(df-mean_val)/s
z.round(2)
```

```python
z.plot(kind='bar')
plt.title("Z scaling Features")
plt.xlabel("Index"); plt.ylabel('z value')
plt.show()
```
